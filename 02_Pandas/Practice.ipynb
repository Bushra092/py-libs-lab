```python
import pandas as pd

data = {
    "Name": ["Amit", "Riya", "Sam", "Neha", "Arjun"],
    "Age": [25, 30, 28, 35, 40],
    "Salary": [40000, 50000, 45000, 60000, 70000],
    "Dept": ["HR", "Finance", "IT", "HR", "IT"]
}

df = pd.DataFrame(data, index =  [101, 102, 103, 104, 105])
df
```

```python
# Practice Questions for You ðŸ”¥

# Using .loc[]:
# Get the Salary of the employee with index 103.
print(f'Get the Salary of the employee with index 103.\n{df.loc[103,'Salary']}')


# Select Name and Age of employees 101 to 104.
print(f'\nSelect Name and Age of employees 101 to 104.\n{df.loc[101:104, ("Name", "Age")]}')

# Using .iloc[]:
# Get the value at first row, second column.
print(f'\nGet the value at first row, second column.\n{df.iloc[1,2]}')

# Select the first 3 rows and first 2 columns.
print(f'\nSelect the first 3 rows and first 2 columns\n{df.iloc[0:3,0:2]}')

# Using Boolean indexing:
# Get all employees with an Age greater than 30.
print(f'\nGet all employees with an Age greater than 30.\n{df[df['Age']>30]}')

# Get employees from the IT department earning more than 45000.
print(f'\nGet employees from the IT department earning more than 45000.\n{df[(df['Dept']=='IT') & (df['Salary'] > 45000)]}')
```

```python


### **Filtering & Sorting**

# 1. Using **`.query()`**, get employees with **Salary > 45000**.
# .query() â†’ filter rows using a string expression
print(f'get employees withSalary > 45000 \n{df.query("Salary > 45000")}')

# 2. Sort the dataset by **Age descending**.
# df.sort_values(by = '', ascending=True/False)
print(f"\nSort dataset by â†’  Age descending\n{df.sort_values(by='Age', ascending = True)}")

# 3. Sort by **Dept first, then Salary ascending**.
print(f"\nSort by Dept first, then Salary ascending\n{df.sort_values(by=['Dept', 'Salary'])}")


### **Aggregation**

# 4. Find the **average Salary per Dept** using `.groupby()`.

print(f"\naverage Salary per Dept\n{df.groupby('Dept')['Salary'].mean()}")

# 5. Using `.agg()`, find **mean, max, min Salary per Dept**.
print(f"\n find â†’ mean, max, min Salary per Dept\n{df.groupby('Dept')['Salary'].agg(["mean","max","min"])}")


# 6. Create a **pivot_table** showing **sum of Salary per Dept**.
pd.pivot_table(df, index = 'Dept', values= 'Salary', aggfunc='sum')
 
```

```python
# Sum of Salary per Dept
print(df.pivot_table(index='Dept', values='Salary', aggfunc='sum'))

# Average Age per Dept
print(df.pivot_table(index='Dept', values='Age', aggfunc='mean'))

# Multiple aggregations
df.pivot_table(index='Dept', values='Salary', aggfunc=['mean','max','min'])
```
