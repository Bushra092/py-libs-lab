```python
import pandas as pd
import numpy as np

# DataFrame 1: Sales Data
sales_data = {
    "OrderID": [101, 102, 103, 104, 105, 106],
    "Customer": ["Alice", "Bob", "Charlie", "Alice", "Eve", "Frank"],
    "Product": ["Laptop", "Tablet", "Laptop", "Phone", "Tablet", "Laptop"],
    "Quantity": [1, 2, 1, 3, 2, 1],
    "Price": [800, 200, 850, 300, 220, 900],
    "OrderDate": pd.to_datetime(["2023-01-01", "2023-01-03", "2023-01-04",
                                  "2023-01-07", "2023-01-10", "2023-01-11"])
}
df_sales = pd.DataFrame(sales_data)

# DataFrame 2: Employee Data
employee_data = {
    "EmpID": [1, 2, 3, 4, 5],
    "Name": ["John", "Sara", "Mike", "Emma", "David"],
    "Department": ["HR", "IT", "Finance", "IT", "HR"],
    "Salary": [50000, 60000, 55000, 65000, 52000],
    "JoiningDate": pd.to_datetime(["2020-02-01", "2019-05-10", "2021-06-20",
                                    "2018-09-15", "2020-12-01"])
}
df_employee = pd.DataFrame(employee_data)

# DataFrame 3: Student Data
student_data = {
    "StudentID": [201, 202, 203, 204, 205, 206],
    "Name": ["Aman", "Priya", "Rohit", "Anita", "Karan", "Simran"],
    "Class": ["10th", "10th", "12th", "11th", "12th", "11th"],
    "Maths": [85, 78, 92, 60, 74, 88],
    "Science": [80, 82, 91, 65, 70, 89],
    "English": [78, 85, 88, 72, 68, 90]
}
df_student = pd.DataFrame(student_data)

(df_sales, df_employee, df_student)
```

```python
#Get the average Price of each product.
df_sales['Price'].mean()
df_sales.groupby('Product')['Price'].mean()
```

```python
# Find the customer who placed the highest order value.

df_sales[df_sales['Revenue'] == df_sales['Revenue'].max()]
```

```python
# Count how many times each Product was sold.
df_sales['Product'].value_counts()
```

```python
# Sort the sales by OrderDate (latest first).
df_sales.sort_values(by = 'OrderDate', ascending = False)
```

```python
# Filter all orders where Price > 500.
df_sales[df_sales['Price'] > 500]
```
