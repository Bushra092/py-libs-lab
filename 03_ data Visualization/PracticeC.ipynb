## ðŸ§  **Practice Task 1 â€” Bar Chart (Sales by Region)**

### **Dataset**

| Region  |  Sales |
| :------ | -----: |
| North   | 150000 |
| South   | 120000 |
| East    | 100000 |
| West    | 180000 |
| Central | 130000 |

### **Tasks**

1. Plot a **vertical bar chart** showing region-wise sales.
2. Use custom colors, e.g. `#5DADE2` (sky blue).
3. Add title **â€œRegional Sales Performanceâ€**.
4. Add X-label = â€œRegionâ€, Y-label = â€œSalesâ€.
5. Add **text labels on top of each bar** showing exact sales values.
6. Add grid lines for better readability.
 

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
region = ['North', 'South', 'East', 'West', 'Central']
sales = [150000, 120000, 100000, 180000, 130000]

df = pd.DataFrame({'Region': region, 'Sales': sales})

df
# 1. Plot a **vertical bar chart** showing region-wise sales.
plt.bar(df['Region'], df['Sales'], color = '#5DADE2')
plt.title('Regional Sales Performance')
plt.xlabel('Region')
plt.ylabel('Sales') 

for i in range(len(df)):
    plt.text(df['Region'][i], df['Sales'][i]+ 5000 , str(df['Sales'][i]) , ha='center', fontsize = 8, fontweight='bold')

plt.ylim(50000, 200000)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show() 
````

### **Dataset**

| Ad_Spend (â‚¹ in 000s) | Sales (â‚¹ in 000s) |
| :------------------: | :---------------: |
|          10          |         25        |
|          15          |         40        |
|          20          |         45        |
|          25          |         60        |
|          30          |         65        |
|          35          |         70        |
|          40          |         85        |

### **Tasks**

1. Plot a **scatter chart** with Ad_Spend on X-axis and Sales on Y-axis.
2. Color points `hotpink`, marker `'o'`, and size `s=100`.
3. Add title: â€œAd Spend vs Sales Relationshipâ€.
4. Add grid lines and axis labels.
5. Try to fit `plt.xlim()` and `plt.ylim()` to make data visible neatly.
 

````python
# Create data
ad_spend = [10, 15, 20, 25, 30, 35, 40]
sales = [25, 40, 45, 60, 65, 70, 85]

# Convert to DataFrame
df_ads = pd.DataFrame({
    'Ad_Spend': ad_spend,
    'Sales': sales
})

plt.scatter(x=df_ads['Ad_Spend'], y=df_ads['Sales'],  color = 'hotpink', marker='o', s=100)
# plt.legend()
plt.title('Ad Spend vs Sales Relationship', fontsize=13, fontweight='bold')
plt.xlabel('Ad Spend')
plt.ylabel('Sales')
plt.xlim((5, 41))
plt.ylim((20,90))

plt.grid()

plt.show() 
````


## **ðŸ“˜ Task 2 â€” Pie Chart (Market Share Distribution)**

```python
companies = ['Apple', 'Samsung', 'Xiaomi', 'OnePlus', 'Others']
market_share = [30, 25, 20, 15, 10]
```

### **Questions**

1. Create a pie chart showing company-wise market share.
2. Add labels and percentage display.
3. Add a starting angle and explode the largest slice.
4. Add title and show the chart neatly.


````python
companies = ['Apple', 'Samsung', 'Xiaomi', 'OnePlus', 'Others']
market_share = [30, 25, 20, 15, 10]

# plt.pie(market_share, labels=companies, autopct='%1.1f%%', startangle=10, explode=(0.05,0,0,0,0))
# plt.title('Market Share Distribution')

explode = [0.1 if val == max(market_share) else 0 for val in market_share]

plt.pie(
    market_share,
    labels=companies,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode,
    shadow=True,
    colors=['#FF9999','#66B3FF','#99FF99','#FFCC99','#C2C2C2']
)

plt.title('Smartphone Market Share', fontsize=13, fontweight='bold')
plt.show()
````


### **Questions**

1. Plot a stacked bar chart showing revenue for Product A and Product B for each quarter.
2. Add legend to distinguish products.
3. Add title and axis labels.
4. Customize colors for each product.
5. Add value labels on each segment if possible.

````python
# quarters = ['Q1', 'Q2', 'Q3', 'Q4']
# product_A = [15000, 18000, 17000, 20000]
# product_B = [12000, 15000, 14000, 16000]

# plt.bar(quarters, product_A, marker='o', label='Product A')

# plt.bar(quarters, product_B , color='#F1948A', marker='^', label='Product B')
# plt.title("Quarterly Revenue Comparison")
# #marker = 'o' (circle), 's' (square), '^' (triangle)
# plt.title("Quarterly Revenue")
# plt.xlabel("product_A")
# plt.ylabel("product_B")
# plt.grid(True)


import matplotlib.pyplot as plt

# Data
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
product_A = [15000, 18000, 17000, 20000]
product_B = [12000, 15000, 14000, 16000]

# Plot
plt.bar(quarters, product_A, color='#5DADE2', label='Product A')
plt.bar(quarters, product_B, bottom=product_A, color='#F1948A', label='Product B')

# Labels & title
plt.title('Quarterly Revenue by Product', fontsize=13, fontweight='bold')
plt.xlabel('Quarter')
plt.ylabel('Revenue')
plt.legend()

# Optional grid and text labels
for i in range(len(quarters)):
    total = product_A[i] + product_B[i]
    plt.text(i, total + 500, str(total), ha='center', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````
