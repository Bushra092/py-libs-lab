### **ðŸ“˜ Q2 â€“ Bar Chart**

**Task:** Compare product sales.
`products = ['A', 'B', 'C', 'D']`, `sales = [150, 200, 120, 300]`

âœ… Make a bar chart.

âœ… Change bar colors.

âœ… Add value labels above bars (hint: use `plt.text()`).

````python
products = ['A', 'B', 'C', 'D']
sales = [150, 200, 120, 300]

plt.bar(products, sales, color = '#546292', edgecolor = 'blue')

for i in range(len(products)):
    plt.text(products[i], sales[i] + 5, str(sales[i]), ha='center', fontsize = 10)
plt.show()
````

### **ðŸ“˜ Q4 â€“ Pie Chart**

**Task:** `categories = ['Rent', 'Food', 'Bills', 'Savings']`, `expenses = [3000, 1500, 1000, 1500]`

âœ… Plot pie chart with labels, autopct, and startangle.

âœ… Explode the â€œRentâ€ section slightly outward.

````python
categories = ['Rent', 'Food', 'Bills', 'Savings']
expenses = [3000, 1500, 1000, 1500]

# Explode the first slice (Rent)
explode_values = [0.05, 0, 0, 0]


plt.pie(x=expenses, labels= categories, startangle=50, autopct='%1.1f%%', explode=explode_values)

plt.title('Monthly Expense Distribution')
plt.show()
````


### **ðŸ“˜ Q5 â€“ Scatter Plot**

âœ… Try adding different colors/sizes for points.

````python
height = [150, 160, 170, 180, 190]
weight = [50, 55, 60, 75, 90]


plt.scatter(height, weight, c=weight, cmap='cool', s=120)
plt.colorbar(label='Weight (kg)')
plt.title('Height vs Weight (Color Mapped)')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()





# plt.scatter(height, weight, alpha=0.8, edgecolor='black')
# plt.scatter(height, weight, alpha=0.8, edgecolor='black')

# plt.title('Height vs Weight')
# plt.xlabel('Height (cm)')
# plt.ylabel('Weight (kg)')
# plt.grid(True)

# plt.show()
````

## ðŸ§  **Practice Task 1 â€” Bar Chart (Sales by Region)**

### **Dataset**

| Region  |  Sales |
| :------ | -----: |
| North   | 150000 |
| South   | 120000 |
| East    | 100000 |
| West    | 180000 |
| Central | 130000 |

### **Tasks**

1. Plot a **vertical bar chart** showing region-wise sales.
2. Use custom colors, e.g. `#5DADE2` (sky blue).
3. Add title **â€œRegional Sales Performanceâ€**.
4. Add X-label = â€œRegionâ€, Y-label = â€œSalesâ€.
5. Add **text labels on top of each bar** showing exact sales values.
6. Add grid lines for better readability.
 

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
region = ['North', 'South', 'East', 'West', 'Central']
sales = [150000, 120000, 100000, 180000, 130000]

df = pd.DataFrame({'Region': region, 'Sales': sales})

df
# 1. Plot a **vertical bar chart** showing region-wise sales.
plt.bar(df['Region'], df['Sales'], color = '#5DADE2')
plt.title('Regional Sales Performance')
plt.xlabel('Region')
plt.ylabel('Sales') 

for i in range(len(df)):
    plt.text(df['Region'][i], df['Sales'][i]+ 5000 , str(df['Sales'][i]) , ha='center', fontsize = 8, fontweight='bold')

plt.ylim(50000, 200000)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show() 
````

### **Dataset**

| Ad_Spend (â‚¹ in 000s) | Sales (â‚¹ in 000s) |
| :------------------: | :---------------: |
|          10          |         25        |
|          15          |         40        |
|          20          |         45        |
|          25          |         60        |
|          30          |         65        |
|          35          |         70        |
|          40          |         85        |

### **Tasks**

1. Plot a **scatter chart** with Ad_Spend on X-axis and Sales on Y-axis.
2. Color points `hotpink`, marker `'o'`, and size `s=100`.
3. Add title: â€œAd Spend vs Sales Relationshipâ€.
4. Add grid lines and axis labels.
5. Try to fit `plt.xlim()` and `plt.ylim()` to make data visible neatly.
 

````python
# Create data
ad_spend = [10, 15, 20, 25, 30, 35, 40]
sales = [25, 40, 45, 60, 65, 70, 85]

# Convert to DataFrame
df_ads = pd.DataFrame({
    'Ad_Spend': ad_spend,
    'Sales': sales
})

plt.scatter(x=df_ads['Ad_Spend'], y=df_ads['Sales'] , color = 'hotpink', marker='o', s=100)
plt.title('Ad Spend vs Sales Relationship', fontsize=13, fontweight='bold')
plt.xlabel('Ad Spend')
plt.ylabel('Sales')
plt.xlim((5, 41))
plt.ylim((20,90))

plt.grid()
plt.show() 
````
