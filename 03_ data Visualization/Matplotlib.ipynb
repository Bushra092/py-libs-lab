### Plotting with Matplotlib and Pandas
Once the DataFrame is ready, various plots can be created using pandas' built-in plotting methods, which utilize matplotlib under the hood. For example, to plot a bar chart, area chart, or line chart:



```python
# df.plot(x = 'year', y='exchange rate', kind='line')
# df.plot(x = 'year', y='exchange rate', kind='bar')
df.plot(x = 'year', y='exchange rate', kind='area')


plt.show()
```

```python
data = { 'Year' : [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020],
        'Exchange Rate': [65,69,71,64,62,59,72,71,75,78,81]
}

df = pd.DataFrame(data)
```

```python
df.plot(x='Year', y= 'Exchange Rate',kind='bar')
```

To create a horizontal bar chart, set kind='barh'.

```python
df.plot(x='Year', y= 'Exchange Rate',kind='barh')
```

### **Scatter Plots**
A scatter plot can be created using the same command by changing kind to 'scatter'.



```python
df.plot(x='Year', y= 'Exchange Rate',kind='scatter')
```

#### Alternatively, matplotlib's scatter method can be used directly:

```python
plt.scatter(x=df['Year'], y= df['Exchange Rate'], color='pink')
```

```python
df.plot(x='Year', y='Exchange Rate', kind='bar')


df.plot(x='Year', y='Exchange Rate', kind='area')


df.plot(x='Year', y='Exchange Rate', kind='line')

```

### **Pie Charts**
To plot a pie chart using pandas:



```python
tasks = {'Pending': 100, 'Completed': 500, 'Ongoing': 300}

df = pd.DataFrame(list(tasks.items()), columns=['Status', 'Tasks'])
df.set_index('Status', inplace=True)
df.plot(y='Tasks', kind='pie')
```

```python

# OR

plt.pie(x=df['Tasks'])
plt.show()
```
