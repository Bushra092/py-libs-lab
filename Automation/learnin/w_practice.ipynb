```python
import openpyxl
from openpyxl import Workbook, load_workbook
```

```python
wb = Workbook()
ws = wb.active
ws.title = "score"
ws.append(['SNo','Name','Score'])
ws.append([101,"Joe", 76])
ws.append([102,"pie", 96])
ws.append([103,"carry", 66])
ws.append([104,"Juliya", 86])
ws.append([105,"James", 85])
ws.append([106,"James", 85])
wb.save("AutomatesFileCreation/practice.xlsx")
```

```python

#? Load `practice.xlsx`, insert an empty row after row 2, save as `practice_v2.xlsx
wb_load = load_workbook("AutomatesFileCreation/practice.xlsx")
ws = wb_load.active
ws.insert_rows(3)
wb_load.save("AutomatesFileCreation/practice_v2.xlsx")
```

```python

# 3. Merge A1:C1, write "CLASS" into merged cell, then unmerge and show values in A1, B1, C1.
wb = load_workbook("AutomatesFileCreation/practice_v2.xlsx")
ws = wb.active
# ws.merge_cells("A1:C1")
# ws['A1'] = "Class"

ws.unmerge_cells("A1:C1")
wb.save("AutomatesFileCreation/practice_v2.xlsx")
```

```python
# 4. Fill a 5×5 block with coordinates (use `get_column_letter`) and save.

from openpyxl.utils import get_column_letter
wb = Workbook()
ws = wb.active

for row in range(1,6):
    for col in range(1,6):
        char = get_column_letter(col)
        # ws[char + str(row)] = char + str(row)
        ws[char+str(row)] = char + str(row)

wb.save("AutomatesFileCreation/dummy_data.xlsx")

```

```python
# 5. Delete the third column in your filled sheet, then print the value that used to be in D1 (now shifted).
wb = load_workbook("AutomatesFileCreation/dummy_data.xlsx")
ws = wb.active

ws.delete_cols(3)

print(ws["D1"].value)
print(ws["D3"].value)
wb.save("AutomatesFileCreation/dummy_data.xlsx")
```
