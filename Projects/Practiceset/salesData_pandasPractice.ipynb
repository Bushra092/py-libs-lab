```python
import pandas as pd
import numpy as np
```

```python
# Load with pd.read_csv()

df = pd.read_csv('sales_data.csv')
df
```

```python
# Filter where Country == "India"

df[df['Country'] == 'India']
```

```python
# Sort by Price

df.sort_values(by='Price', ascending=False)
```

```python
# Group by Category and find total sales
df.groupby('Category')['Price'].sum()
```

```python
# Handle missing Quantity and Price with 
AvgPRice = df['Price'].mean()

print(AvgPRice)
print(df['Price'].fillna(AvgPRice, inplace=True))

df
```

```python
# Handle missing Quantity 
AvgQunaity = df['Quantity'].mean()

print(AvgQunaity)
df['Quantity'].fillna(AvgQunaity, inplace=True)

df
```
