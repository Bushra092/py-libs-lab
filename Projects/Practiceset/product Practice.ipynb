```python
import pandas as pd
```

```python
df = pd.read_csv("product_feedback.csv")
df
```

```python

# Use .query('Rating >= 4')
df.query('Rating >= 4')

# Sort by Rating
df.sort_values(by='Rating', ascending=False)

# Group by Country to get average rating
df.groupby('Country')['Rating'].mean()

# Handle missing Rating with .fillna()
avgrating = df['Rating'].mean()
df.fillna(avgrating, inplace= True)
df

# Drop irrelevant column (e.g. .drop('Feedback', axis=1))
df.drop('Feedback', axis=1)
```
