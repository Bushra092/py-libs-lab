

#### ðŸ§¹ **Cleaning Data â€” `.dropna()`, `.fillna()`, `.replace()`, `.drop()`**

16. Find how many rows have missing values.
17. Replace missing discounts with `0.0` using `.fillna()`.
18. Drop all rows where `Profit` is missing.
19. Replace the word `"Accessories"` in the `Category` column with `"Gadgets"`.
20. Drop the `Date` column using `.drop()`.



```python
# 16. Find how many rows have missing values.
df.isnull().sum()

# 17. Replace missing discounts with `0.0` using `.fillna()`.
df['Discount'].fillna(0.0, inplace=True)
df

# 18. Drop all rows where `Profit` is missing.
df.dropna(inplace=True)
df

# 19. Replace the word `"Accessories"` in the `Category` column with `"Gadgets"`.
df.replace({"Accessories": "Gadgets"}, inplace=True)
df
 

# 20. Drop the `Date` column using `.drop()`.
df.drop('Date', axis=1, inplace=True)
df
```



#### ðŸ“ˆ **Visualization â€” `.plot()`**

21. Plot total `Sales` per `Region` (bar chart).
22. Plot `Sales` vs. `Profit` (scatter).
23. Create a line chart showing `Sales` over `Date`.


```python
# 21. Plot total `Sales` per `Region` (bar chart).
df.groupby(by = 'Region')['Sales'].sum().plot(kind='bar', color='#a06050', title='Total Sales by Region')

# 22. Plot `Sales` vs. `Profit` (scatter).
df.plot(x = 'Sales', y='Profit', kind= 'scatter', color ='hotpink', colorbar='pink')

# 23. Create a line chart showing `Sales` over `Date`.
df.plot(x='Date', y= 'Sales', kind= 'line')

```
