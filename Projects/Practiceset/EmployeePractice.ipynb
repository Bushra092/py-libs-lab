```python
import pandas as pd
```

```python
# ‚úÖ Practice ideas:
df = pd.read_csv('employee_data.csv')
print(df)


# Select rows using .loc[] and .iloc[]
df.iloc[[0,2,4]]
# df.loc[6]



# Replace missing City with ‚ÄúUnknown‚Äù
# df['City'].fillna('‚ÄúUnknown‚Äù', inplace=True) 
# df

# Drop rows where Salary is missing
df.dropna(subset=['Salary'], inplace=True)
df 
# Filter employees with Salary > 60000
df[df['Salary'] > 60000]


# Create a Level column using conditions (Junior, Mid, Senior)
df['Experience'].fillna(0, inplace=True) 
df

df['Level'] = df['Experience'].apply(
    lambda x: 'Junior' if x<2 else 'Mid' if x < 4 else 'Senior'
)

df
```

```python
# create a column for salary grade‚Äù (Low, Medium, High)

df['ExpSalary'] = df['Salary'].apply(
    lambda x: 'Low' if x<60000 else 'Medium' if x<= 70000 else 'High' 
)
df
```
