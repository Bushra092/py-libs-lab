# **Probability for Analysis**

## ⚙ Practice

1. Assume 1,000 users visited your website
2. 300 clicked button
3. 100 added to cart
4. 50 purchased

Do this
• P(click)

• P(add to cart)

• P(purchase | add to cart)

• P(click or add to cart)


```python
import pandas as pd

df = pd.read_csv("website_actions.csv")
```

```python
print(df.head(3))
print(df.shape)
  
```

```python
p_cart = df['added_to_cart'].sum()
p_cliks = df['clicked'].sum()
p_purchased = df['purchased'].sum()
total = df.shape[0]
print(total,p_cart,p_cliks,p_purchased)

# Do this
# • P(click)
print("\n------P(click)----- ")
print(p_cliks/total) # 3/10


# • P(add to cart)

print("\n------P(add to cart)----- ")
print(p_cart/total) # 1/10

# • P(purchase | add to cart)
print("------P(purchase | add to cart)-----\n ")
purchased_Cart = df[(df['added_to_cart'] == 1) & (df['purchased'] == 1)].shape[0]
print(purchased_Cart)
print(purchased_Cart/p_cart)  # 1/20

# • P(click or add to cart)
print("------ P(click or add to cart)----- ")

clicked_and_cart = df[(df['clicked']==1) & (df['added_to_cart']==1)].shape[0] 
print((p_cliks/total) + (p_cart / total) - (clicked_and_cart / total))
```
